### Importing libraries

```python
import pandas as pd
```

### Importing the data frame

```python
df = pd.read_csv("Pizza_sales.csv", index_col=0)
df
```

### Examining the data frame

```python
df.dtypes
```

### Step 01

Converting data types

```python
df
```

```python
# Convert Order_Date to datetime
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='mixed', errors='raise', dayfirst=True)
```

```python
# Convert Sales to numeric
df['Sales'] = df['Sales'].astype(float)
```

```python
# 3. Convert to categorical
df['Pizza_Type'] = df['Pizza_Type'].astype('category')
df['Size'] = df['Size'].astype('category')
df['Branch'] = df['Branch'].astype('category')
```

```python
df
```

```python
df.dtypes
```

Standardizing text values

```python
df['Pizza_Type'] = df['Pizza_Type'].str.title()
df['Branch'] = df['Branch'].str.upper()
```

```python
df
```

```python
df['Size'].value_counts()
```

Remapping size

```python
df['Size'] = df['Size'].map({'S': 'Small', 'M': 'Medium', 'L': 'Large'})
```

```python
df
```

### Step 02

```python
# Renaming columns 
df.rename(columns={'Pizza_Type': 'Topping', 'Order_Date': 'Date'}, inplace=True)
```

```python
df
```

```python
# Reordering the columns
df = df[['Order_ID', 'Date', 'Customer', 'Topping', 'Size', 'Sales', 'Branch']]
```

```python
#Setting up index
df.set_index('Order_ID', inplace=True)
df.head()

df
```

```python
# Reset later if needed
df.reset_index(inplace=True)
```

```python
df
```

### Step 03

```python
# Simple grouping
df.groupby('Branch')['Sales'].sum()
```

```python
# Grouping by multiple variables
df.groupby(['Branch', 'Topping'])['Sales'].mean().reset_index()
```

```python
df.groupby('Branch').agg({
    'Sales': ['sum', 'mean', 'count']
})
```

```python
df.groupby('Branch').agg(
    Total_Sales=('Sales', 'sum'),
    Avg_Sale=('Sales', 'mean'),
    Num_Orders=('Sales', 'count')
).reset_index()
```

### Step 04

```python
# Using pivot_table to handle duplicates
pivot_df = df.pivot_table(
    index='Branch',      # rows
    columns='Topping',   # columns
    values='Sales',      # values to fill
    aggfunc='sum',       # aggregation function for duplicates
    fill_value=0         # optional, fill missing values with 0
)

pivot_df
```

```python
# Melted (long) data frame
melt_df = pd.melt(pivot_df.reset_index(),
                  id_vars=['Branch'],
                  var_name='Topping',
                  value_name='Sales')
melt_df.sort_values(by="Sales", ascending=False)
```

```python
# Stacked data frame
stacked = pivot_df.stack()
stacked
```

```python
# Unstacking again
unstacked = stacked.unstack()
unstacked
```
