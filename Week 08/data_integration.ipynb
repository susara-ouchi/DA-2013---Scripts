```python
import pandas as pd
```

```python
pizza_orders = pd.read_csv("Pizza_orders.csv", index_col=0)
pizza_orders
```

```python
pizza_prices = pd.read_csv("pizza_prices.csv", index_col=0)
pizza_prices
```

```python
# Merge tables to get Price 
pizza_sales = pizza_orders.merge(pizza_prices, on=['Pizza_Type', 'Size'], how='left')

# Compute actual Sales
pizza_sales['Sales'] = pizza_sales['Quantity'] * pizza_sales['Price']

pizza_sales
```

```python
pizza_sales.to_csv("Pizza_sales.csv")
```
